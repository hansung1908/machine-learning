```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# MNIST(필기체) 데이터셋을 로드합니다
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# 데이터 전처리
# 이미지 데이터를 28 x 28의 픽셀값으로 처리(처리하는 사진 숫자, 픽셀수)
train_images = train_images.reshape((60000, 28 * 28))
test_images = test_images.reshape((10000, 28 * 28))
# 각 픽셀값은 색상값으로 0~255의 정수로 표현
train_images = train_images.astype('float32') / 255
test_images = test_images.astype('float32') / 255
# 라벨 데이터를 배열값으로 변환
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)

# 신경망 모델 구성
model = models.Sequential(name = 'mnist_net')
model.add(layers.Dense(512, activation = 'relu', input_shape=(28 * 28,), name = 'input_dense')) # 전처리한 픽셀 값 데이터 입력, 512개의 뉴런
model.add(layers.Dropout(0.5, name = 'hidden_layer')) # 과적합을 방지하기 위한 dropout 레이어
model.add(layers.Dense(10, activation='softmax', name = 'output_dense')) # 10개의 종류중 하나의 속할 확률 출력

# 모델 컴파일
model.compile(optimizer='adam',
              loss='categorical_crossentropy', # 다중 분류
              metrics=['accuracy'])

# 모델 훈련
model.fit(train_images, train_labels, epochs=5, batch_size=64)

# 모델 평가
test_loss, test_acc = model.evaluate(test_images, test_labels) # 손실도, 정확도
print(f'Test accuracy: {test_acc}')
```
