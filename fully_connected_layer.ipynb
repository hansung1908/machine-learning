```python
from threading import activeCount
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical

# dataset으로 부터 붓꽃 데이터 불러오
iris = load_iris()

data = iris['data'] # 특징 값(꽃잎과 꽃받침의 너비와 길이값)
answer = iris['target'] # 정답(0, 1, 2)
print(iris.keys())
```

```python
# 학습 특징데이터, 평가 특징데이터, 학습 정답데이터, 평가 정답데이터
# 20%를 평가 셋으로 지정하는 경 값은 '0.2'
x_train, x_test, y_train, y_test = train_test_split(data, answer, test_size = 0.2)

print(x_train.shape)
print(y_train[:10])
```

```python
# 붓꽃 데이터는 3가지의 결과 중 유사 값이 결정되어야 함으로 정답 값은 [0, 0, 0] 형태를 가지는 1차원 배열
# to_categorical 함수를 사용하여 정답 값을 3의 길이를 가진 1차원 배열로 반환
y_train = to_categorical(y_train, 3)
y_test = to_categorical(y_test, 3)

print(y_test.shape)
print(y_train.shape)
print(y_train[:10])
```

```python
model = Sequential(name = 'nsu_irisnet') # nsu_irisner이라는 이름을 가진 학습 모델 생성
model.add(Dense(128, activation = 'relu', input_shape = (4,), name = 'input_dense')) # input layer
model.add(Dense(64, activation = 'relu', name = 'hidden_dense1')) # hidden layer1
model.add(Dense(32, activation = 'relu', name = 'hidden_dense2')) # hidden layer2
model.add(Dense(3, activation = 'softmax', name = 'output_dense')) # ouput layer, 붓꽃의 종류는 3종류로 최종 뉴런은 3개

# 붓꽃 데이터의 경우, 3가지 종류 중 한 종류가 정답이 되기에 다중 분류에 속함, categorical_crossentropy
model.compile(optimizer = "adam", loss = "categorical_crossentropy", metrics = ['accuracy'])

# fit 함수를 통해 학습 시작, epochs는 학습 반복
hist = model.fit(x_train, y_train, epochs = 40)
```

```python
# evaluate 함수로 별도로 분리 시켜놓았던 평가 셋의 데이터를 입력, 정확도는 약 96%
result = model.evaluate(x_test, y_test, batch_size=32)
```
